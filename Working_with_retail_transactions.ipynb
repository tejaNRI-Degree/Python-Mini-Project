```python
#Data Loading
df=pd.read_csv('/content/drive/MyDrive/retail transaction ')
```

```python
# TransactionDate column is parsed into datetime object
df['TransactionDate']=pd.to_datetime(df['TransactionDate'])
```

```python
print ("First 15 rows of the DataFrame:\n")
print(df.head(15))
```

```python
print("Summary of the DataFrame:\n")
print(df.info())
```

```python
print("Descriptive Statistics:\n")
df.describe()
```

```python
df.count()
```

```python
missingValues=df.isnull().sum()
```

```python
print("Missing values in each column:\n")
print(missingValues)
```

```python
# For the columns Quantity, Price, and Discount, ensuring that they are in the correct numerical format.
df['Quantity'] = pd.to_numeric(df['Quantity'])
```

```python
df['Quantity']
```

```python
df['Price'] = pd.to_numeric(df['Price'])
```

```python
df['Price']
```

```python
df['Discount'] = pd.to_numeric(df['Discount'])
```

```python
df['Discount']
```

```python
df['Discount'] = df['Discount'].fillna(0)
```

```python
print("Filling missing values in the Discount column with 0:\n")
df['Discount']
```

```python
# Creating a new column called NetPrice which applies the discount to the price
df['NetPrice'] = df['Price'] * (1 - df['Discount'] / 100)
```

```python
df['NetPrice']
```

```python
df['TotalSale'] = df['Quantity'] * df['NetPrice']
```

```python
df['TotalSale']
```

```python
print("DataFrame with new NetPrice and TotalSale columns:\n")
print(df.head())
```

```python
# Data Filtering
# Filtering the DataFrame that include only those transactions where TotalSale is greater than or equal to 20.
filter1=df[df['TotalSale'] >= 20]
```

```python
filter1
```

```python
# Converting DataType of PaymentType into Categorical DataType
df['PaymentType']=df['PaymentType'].astype('category')
```

```python
print("DataType after converting PaymentType to category:\n")
print(df.dtypes)
```

```python
# Grouping the filtered DataFrame by ProductCategory and computing total Quantity sold for each category
filter2 = df.groupby('ProductCategory').agg(TotalQuantity=('Quantity','sum'),TotalSale=('TotalSale','sum'))
```

```python
filter2
```

```python
# Sorting the grouped result in descending order based on the total TotalSale.
sorted_group=filter2.sort_values(by='TotalSale', ascending=False)
```

```python
sorted_group
```

```python
# Identifing and printing the top 3 product categories with the highest total sales.
top_3_categories = sorted_group.head(3)
```

```python
print("Top 3 product categories with the highest total sales:\n")
print(top_3_categories)
```

```python
# Multi-Level Grouping
# Grouping the filtered data by both StoreID and ProductCategory
grouped_data = df.groupby(['StoreID', 'ProductCategory'])['TotalSale'].mean().reset_index()
```

```python
grouped_data
```

```python
# For each group, calculating the average TotalSale per transaction.
grouped_data=grouped_data.rename(columns={'TotalSale':'AvgTotalSale'})
```

```python
print("Multi-level grouping by StoreID and ProductCategory with average TotalSale:\n")
print(grouped_data)
```

```python
# For each store (StoreID), determining which product category has the highest average TotalSale.
top_product_categories=grouped_data.loc[grouped_data.groupby('StoreID')['AvgTotalSale'].idxmax()]
```

```python
print(top_product_categories)
```

```python
summary_df=top_product_categories[['StoreID','ProductCategory','AvgTotalSale']]
print("Store-Level Insights:\n",summary_df)
```
